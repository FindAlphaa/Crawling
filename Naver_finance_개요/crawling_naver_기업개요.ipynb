```python
# Create a dictionary mapping company names to their corresponding 종목코드
company_to_code = {row['회사명']: row['종목코드'] for _, row in code_df.iterrows()}

# Add a new column '종목코드' to df based on the mapping
df['ticker'] = df['Company'].map(company_to_code)
print(df)
# Save the updated df to a new CSV file
df.to_csv("updated_software.csv", index=False, encoding="cp949")
```
