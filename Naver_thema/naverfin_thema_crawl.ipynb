# 네이버 금융 테마 크롤링

## Import

```python
import requests               
from bs4 import BeautifulSoup
import lxml
from collections import defaultdict
import pandas as pd
```

## Thema 

```python
thema_dict = defaultdict(list)

for i in range(1,8):
    url = f'https://finance.naver.com/sise/theme.naver?&page={i}'

    req = requests.get(url, headers = {'User_agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"})
    html = req.text

    soup = BeautifulSoup(html, 'lxml')
    #print(soup)
    soup.find_all('.col_type1')

    thema = soup.find_all('td','col_type1')
    for t in thema:
        thema_dict['thema'].append(t.text)
print(thema_dict['thema'])
```

```python
thema = pd.DataFrame(thema_dict)
thema.to_csv("thema.csv", index = False)
```
